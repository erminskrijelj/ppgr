```python
def jednacine(l,r):

    tmp1 = np.array([l[1]*T1[2]-l[2]*T1[1],
                    -l[0]*T1[2]+l[2]*T1[0],
                     r[1]*T2[2]-r[2]*T2[1],
                    -r[0]*T2[2]+r[2]*T2[0]])
    return np.matrix(tmp1)
```

```python
def tridkoordinate(l,r):
    mat = np.transpose(np.linalg.svd(jednacine(l, r))[2][3])
    koord = np.array(mat)
    koord = koord / koord[3]
    return koord[:-1]
```
